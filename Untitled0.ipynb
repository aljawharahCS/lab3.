```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Test_Data.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.hist(df['age'])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df['bmi'])
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.countplot(x='sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='smoker', data=df)
plt.title("Smoker Distribution")
plt.show()
```

```python
sns.countplot(x='region', data=df)
plt.title("Region Distribution")
plt.show()
```

```python
sns.scatterplot(x='age', y='bmi', data=df)
plt.title("Age vs BMI")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
## Key Insights

1. Most individuals are non-smokers.
2. Age ranges between 18 and 64 years.
3. BMI values are centered around 30.
4. Most individuals have between 0â€“3 children.
5. No strong correlation exists between numerical variables.
6. Gender distribution appears balanced.
```
